### Mengatasi missing value pada katagori colum


```python
missing_categorical = crypto.select_dtypes(include=['object']).isnull().sum()

missing_categorical = missing_categorical[missing_categorical > 0]

print(missing_categorical)
```

Disini saya mengatasinya dengan metode modus, yaitu mengisi nilai kosong dengan nilai yang sering keluar.

```python
for col in ['platform_name', 'platform_symbol', 'platform_slug', 'platform_token_address']:
    crypto[col].fillna(crypto[col].mode()[0], inplace=True)
```

```python
print(crypto.isnull().sum())
```

Bisa Dilihat Data yang katagorikal tadi sudah bersih dilihat dari data di atas

### Mengatasi missing value pada numeric colum


```python
missing_categorical = crypto.select_dtypes(include=['number']).isnull().sum()

missing_categorical = missing_categorical[missing_categorical > 0]

print(missing_categorical)
```

Disini karena data bertype numeric saya akan menggunakan metode median yaitu mengisi nilai yang kosong dengan nilai median dari column tersebut

```python
for col in ['max_supply', 'self_reported_circulating_supply', 'self_reported_market_cap', 'platform_id']:
    crypto[col].fillna(crypto[col].median(), inplace=True)
```

```python
print(crypto.isnull().sum())
```
